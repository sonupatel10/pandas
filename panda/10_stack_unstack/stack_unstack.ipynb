```python
import pandas as pd
```

```python
df = pd.read_excel("stocks.xlsx",header=[0,1],index_col=0)
df
```

```python
new_df = df.stack(future_stack=True)

new_df
```

```python
new_df.unstack()
```

```python
df1=pd.read_excel('stocks_3_levels.xlsx',header=[0,1,2],index_col=0)
df1
```

```python
df_stack=df1.stack()
df_stack
```

```python
df_stack=df1.stack(level=1)
df_stack
```
