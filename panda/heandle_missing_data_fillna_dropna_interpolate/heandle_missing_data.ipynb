```python
df=pd.read_csv('weather_data.csv',parse_dates=['day'])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'not event'
    
})
new_df
```

```python
# new_df=df.fillna(method='ffill') #-> old syntax
new_df=df.ffill()# -> new syntax
new_df
```

```python
# new_df=df.fillna(method='bfill') #-> old sysntx
new_df=df.bfill() #new syntsx
new_df
```

```python
#new_df=df.fillna(method='ffill',limit=1)
new_df=df.ffill(limit=1)
new_df
```

```python
new_df=df.ffill(limit=2)
new_df
```

```python
new_df=df.ffill(axis='columns')
new_df
```

```python
new_df=df.bfill(axis='columns')
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how='all')
new_df
```

```python
new_df=df.dropna(thresh=1)
new_df
```

```python
new_df=df.dropna(thresh=1)
new_df
```
