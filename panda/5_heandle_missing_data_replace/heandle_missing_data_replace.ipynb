```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('weather_data (1).csv')
df
```

```python
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
new_df=df.replace([-88888,-99999],np.NaN)
new_df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888]
    
},np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
     '0':'sunny'
    
})
new_df
```

```python
new_df=df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
new_df=df.replace({
    'windspeed':'[A-Za-z]',
    'temperature':'[A-Za-z]'	
                  
},' ',regex=True)
new_df
```
