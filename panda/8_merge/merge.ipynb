```python
import pandas as pd
```

```python
df1=pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df1=pd.DataFrame({
    "city": ["new york","chicago","orlando","belmati"],
    "temperature": [21,14,35,44],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando","senfrancico"],
    "humidity": [65,68,75,22],
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='inner')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='left')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='right')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='right',indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city")
df3
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second')) # replace _x,_y to _first,_second
df3
```
