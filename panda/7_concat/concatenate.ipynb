```python
import pandas as pd
```

```python
india_weather=pd.DataFrame({
    'city':['mumbai','surat','delhi'],
    'humidity':[34,56,87],
    'temperature':[23,44,56]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    'city':['new yourk','chicago','adelaide'],
    'humidity':[45,66,99],
    'temperature':[67,53,26]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,us_weather],keys=['india','us'])
df
```

```python
df.loc['india']
```

```python
df.loc['us']
```

```python
temperature_df=pd.DataFrame({
    'city':['mumbai','surat','delhi'],
    'temperature':[12,45,86]
    
})
temperature_df
```

```python
windspeed_df=pd.DataFrame({
    'city':['mumbai','surat','delhi'],
    'windspeed':[67,43,88]
    
})
windspeed_df
```

```python
df=pd.concat([temperature_df,windspeed_df])
df
```

```python
df=pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
temperature_df=pd.DataFrame({
    'city':['mumbai','surat','delhi'],
    'temperature':[12,45,86]
    
},index=[0,1,2])
temperature_df
```

```python
windspeed_df=pd.DataFrame({
    'city':['delhi','mumbai'],
    'windspeed':[67,43]
    
},index=[2,0])
windspeed_df
```

```python
df=pd.concat([temperature_df,windspeed_df],axis=1)
df
```
