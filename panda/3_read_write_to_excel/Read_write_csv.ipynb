```python
#use when your dataset have no header and give header name , butmy data set have header so it show as row
df=pd.read_csv('stock_data.csv',header=None,names=['ticker','eps','revenue','price','people'])
df
```

```python
#if you want need only some number of rows use it
df=pd.read_csv('stock_data.csv',nrows=3)
df
```

```python
df=pd.read_csv('stock_data.csv',na_values=['not available','n.a.'])
df
```

```python
df=pd.read_csv('stock_data.csv',na_values={
    'eps':['n.a.','not available'],
    'revenue':['not avilable','n.a.',-1],
    'price':['not avilable','n.a.'],
    'people':['not avilable','n.a.',-1]
    
})
df
```

```python
#if you don't need first index use index=Flase
# if you don
df.to_csv('new.csv')
# df.to_csv('new.csv',index=Flase)
```

```python
#if you want to export few columns use this
df.to_csv('new.csv',columns=['eps','people'])
```

```python
# change specific values
import pandas as pd
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'sonu patel'
    return cell
df=pd.read_csv('stock_data.csv',converters={
        'people':convert_people_cell
    })
df
```
