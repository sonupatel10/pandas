```python
import pandas as pd
```

```python
df=pd.read_csv('weather_data.csv')
df
```

```python
df=pd.read_excel('weather_data.xlsx','Sheet1')
df
  
```

```python
weather_data={
    'day':['1/1/2025','2/1/2025','3/1/2025'],
    'temp':[20,45,67],
    'windspeed':[32,56,90],
    'event':['Rain','sunny','cloud']
}
df=pd.DataFrame(weather_data)
df
```

```python
weather_data={
    ('1/1/2025',20,32,'Rain'),
    ('2/1/2025',45,67,'sunny'),
    ('3/1/2025',56,90,'cloud'),
    
}
df=pd.DataFrame(weather_data,columns=['day','temp','windspeed','event'])
df
```

```python
weather_data=[
    {'day':'1/1/2025','temp':20,'windspeed':56,'event':'Rain'},
    {'day':'2/1/2025','temp':67,'windspeed':43,'event':'sunny'},
    {'day':'3/1/2025','temp':80,'windspeed':23,'event':'cloud'}
    ]
     
df=pd.DataFrame(weather_data)
df
```
